```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
```

# 预测国际航班未来一个月的乘客数

```python
# 加载数据
data = pd.read_csv('../data/international-airline-passengers.csv', usecols=[1])
plt.plot(data.values.astype('float32'))

train = data['passengers'][0:120].values.astype('float32')
test = data['passengers'][120:].values.astype('float32')
```

```python
# 拟合模型并做预测
model = auto_arima(train, error_action='ignore', m=12)
preds, conf_int = model.predict(n_periods=test.shape[0], return_conf_int=True)
print("Test RMSE: %.3f" % np.sqrt(mean_squared_error(test, preds)))
plt.plot(preds)
plt.plot(test)
```

# 脱敏后的采购数据做预测

```python
data2 = pd.read_excel('../data/data_20190203.xlsx', sheet_name='Sheet1')

# 预测未来10天的采购量
train = np.log1p(data2['apply_num'][0:120]).values.astype('float32')
test = np.log1p(data2['apply_num'][120:130]).values.astype('float32')

modl = auto_arima(train, error_action='ignore')
preds, conf_int = modl.predict(n_periods=test.shape[0], return_conf_int=True)
print("Test RMSE: %.3f" % np.sqrt(mean_squared_error(test, preds)))

plt.plot(preds)
plt.plot(test)
```
