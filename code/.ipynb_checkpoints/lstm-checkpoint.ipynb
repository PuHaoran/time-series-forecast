```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import read_csv
import math
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Activation
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

```python
# 加载数据
data = pd.read_csv('../data/international-airline-passengers.csv', usecols=[1])
```

```python
data = data.values
data = data.astype('float32')
data.shape
```

```python
# 12年，每年12个月数据，共144个数据
plt.plot(data)
```

# 数据预处理
    

```python
def create_dataset(dataset, look_back=1):
    """ look_back是预测下一步所需要的时间不长 """
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i: (i + look_back)]
        dataX.append(a)
        dataY.append(dataset[i + look_back])
    
    return np.array(dataX), np.array(dataY)

np.random.seed(7)
```

```python
# 正则化数据集
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(data)

# 划分测试集和验证集
train_size = int(len(data) * 0.67)
train, test = data[0:train_size], data[train_size:len(data)]
```

```python
look_back = 10
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
```

```python
print('trainX.shape', trainX.shape)
print('trainY.shape', trainY.shape)
```

```python
# LSTM的输入需要这样的结构 [samples, timesteps, features]
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))
```

```python
# 建立LSTM模型
"""
model = Sequential()
model.add(LSTM(4))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
"""
```

```python
model = Sequential()
model.add(LSTM(input_dim=1, output_dim=50, return_sequences=True, dropout=0.2))
model.add(LSTM(100, return_sequences=False, dropout=0.2))
model.add(Dense(output_dim=1))
model.add(Activation('linear'))
model.compile(loss='mse', optimizer='rmsprop')
```

```python
model.fit(trainX, trainY, epochs=20, batch_size=1, verbose=2)
```

```python
# 预测
train_pred = model.predict(trainX)
test_pred = model.predict(testX)
```

```python
# 转换数据为相同单位，计算误差
train_pred = scaler.inverse_transform(train_pred)
trainY = scaler.inverse_transform(trainY)
test_pred = scaler.inverse_transform(test_pred)
testY = scaler.inverse_transform(testY)
```

```python
# mse
train_score = math.sqrt(mean_squared_error(trainY[:, 0], train_pred[:, 0]))
print('训练集分数 %.2f RMSE' % (train_score))

test_score = math.sqrt(mean_squared_error(testY[:, 0], test_pred[:]))
print('测试集分数 %.2f RMSE' % (test_score))
```

```python
# 测试集结果
# 蓝色为原数据，红色为预测数据
plt.plot(testY)
plt.plot(test_pred)
```

```python
# 验证集结果
plt.plot(trainY)
plt.plot(train_pred)
```
